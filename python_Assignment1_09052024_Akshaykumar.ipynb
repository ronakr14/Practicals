Q1. Plot secondary axis in matplotlib with legends

```python
import matplotlib.pyplot as plt
 
# Sample data for Plot A (assuming x-axis is same for both plots)
x_data = [10, 20, 30, 40, 50]
y1_data = [42, 55, 70, 85, 30]
 
# Sample data for Plot B
y2_data = [51, 42, 68, 72, 70]
 
y3_data = [1, 14, 28, 12, 8]
y4_data = [82, 23, 78, 4, 90]

# Create the figure and subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

# Plot data from Plot A and Plot B
ax = axes[0]
 
# Plot data from Plot A
bar1 = ax.plot(x_data, y1_data, label='Plot A', color='red')
 
# Create a twin axis for Plot B
ax2 = ax.twinx()
 
# Plot data from Plot B on the secondary axis
bar2 = ax2.plot(x_data, y2_data, label='Plot B', color='orange')
 
# Set y-axis limits for the merged plot
ax.set_ylim(0, 100)  
ax2.set_ylim(0, 100) 
 
# Set labels and title for the merged plot
plt.xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label for Plot A') 
ax2.set_ylabel('Y-axis Label for Plot B') 
 
# Add legend for the merged plot
handles1, labels1 = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
plt.legend(handles1 + handles2, labels1 + labels2, loc='upper left')  
 
# Plot data from Plot C and Plot D
ax3 = axes[1]

# Plot data from Plot C on the new axes
bar3 = ax3.plot(x_data, y3_data, label='Plot C', color='yellow')
 
# Create a twin axis for Plot D on the new axes
ax4 = ax3.twinx() 
 
# Plot data from Plot D on the secondary axis
bar4 = ax4.plot(x_data, y4_data, label='Plot D', color='green')
 
# Set y-axis limits for the second plot
ax3.set_ylim(0, 100) 
ax4.set_ylim(0, 100) 
 
# Set labels and title for the second plot
plt.xlabel('X-axis Label')
ax3.set_ylabel('Y-axis Label for Plot C') 
ax4.set_ylabel('Y-axis Label for Plot D') 
 
# Add legend for the second plot
handles3, labels3 = ax3.get_legend_handles_labels()
handles4, labels4 = ax4.get_legend_handles_labels()
plt.legend(handles3 + handles4, labels3 + labels4, loc='upper left')
 
# Create the figure and subplot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot data from Plot A and Plot C
ax.plot(x_data, y1_data, label='Plot A', color='red')
ax.plot(x_data, y3_data, label='Plot C', color='yellow')

# Set labels and title for the plot
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y1,Y3-axis Label')
ax.set_title('Combined Plot') 

# Create a twin axis for Plot B and Plot D
ax2 = ax.twinx()

# Plot data from Plot B and Plot D on the secondary axis
ax2.plot(x_data, y2_data, label='Plot B', color='orange')
ax2.plot(x_data, y4_data, label='Plot D', color='green')

# Set labels for the secondary axis
ax2.set_ylabel('Y2,Y4-axis Label')

# Add legends
handles1, labels1 = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax.legend(handles1 + handles2, labels1 + labels2, loc='upper left')
 
plt.tight_layout()
# Show the plots
plt.show()
```

Q2. Use lambda operations on the dataframes

```python
import pandas as pd
data = {
    'A': [1, 2, 3, 4],
    'B': [11,22,33,44],
}
df = pd.DataFrame(data)
df['A_+_B'] = df.apply(lambda row: row['A'] + row['B'], axis=1)
df['A_sqr'] = df['A'].apply(lambda x: x ** 2)
df['B_mul'] = df['B'].map(lambda x: x * 5)
df
```

Q3. Python program to check anagrams

```python
def anagrams_word(word1, word2):
    word1 = word1.lower().replace(" ", "")
    word2 = word2.lower().replace(" ", "")
    
    if sorted(word1) == sorted(word2):
        print(f"Both strings {word1} and {word2} are anagrams")
    else:
        print(f"Both strings {word1} and {word2} are not anagrams")

w1 = "elbow"
w2 = "below"
w3="cpp"
w4="java"

anagrams_word(w1, w2)
anagrams_word(w3, w4)
```

Q5. Explain Types of merges?

- Types of Git Merges:
1) Normal Merge:
- It creates new merge commit that combines the changes from both branches head branch and target branch.
- Git performs a three-way comparison to identify any conflicts.
- If there are conflicts, you'll need to manually resolve them before the merge is complete.
- This approach preserves the complete Git history, showing the evolution of both branches and the merge point.

2) Squash Merge:
- It's similar to a normal merge, it combines changes from branches.
- Instead of creating a new merge commit, it condenses the commits from the head branch into a single new commit on the target branch.
- This keeps the history cleaner for branches with many small commits.
- Use this when you want a simpler history for feature branches or bug fixes that don't need a detailed record of individual commits.

3) Rebase Merge:
- It replays the commits from the head branch on top of the target branch.
- If there are no conflicts, Git integrates the changes and updates the branch pointer to the last commit of the head branch.
- Use this for feature branches that are up-to-date with the target branch and don't require preserving the individual commits in the history.
