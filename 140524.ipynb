```python
import matplotlib.pyplot as plt

# Sample data for Plot A (assuming x-axis is same for both plots)
x_data = [1, 2, 3, 4, 5]
y1_data = [10, 15, 20, 25, 30]

# Sample data for Plot B
y2_data = [38, 13, 27, 22, 30]
y3_data = [25, 29, 14, 23, 30]
y4_data = [20, 21, 13, 24, 30]

# Create the figure and subplots
fig, ax = plt.subplots()

# Plot data from Plot A
bar1 = ax.plot(x_data, y1_data, label='Plot A', color='red')

# Create a twin axis for Plot B
ax2 = ax.twinx()  # Creates a new axes object with same x-axis

# Plot data from Plot B on the secondary axis (use bottom argument to stack bars)
bar2 = ax2.plot(x_data, y2_data, label='Plot B', color='orange')  # Different color for clarity

# Create a twin axis for Plot C
ax3 = ax.twinx()  # Creates a new axes object with same x-axis

# Plot data from Plot C on the secondary axis (use bottom argument to stack bars)
bar3 = ax3.plot(x_data, y3_data, label='Plot C', color='yellow')  # Different color for clarity

# Create a twin axis for Plot D
ax4 = ax.twinx()  # Creates a new axes object with same x-axis

# Plot data from Plot D on the secondary axis (use bottom argument to stack bars)
bar4 = ax4.plot(x_data, y4_data, label='Plot D', color='green')  # Different color for clarity

# Align y-axis ticks
y_ticks = ax.get_yticks()
ax2.set_yticks(y_ticks)
ax3.set_yticks(y_ticks)
ax4.set_yticks(y_ticks)

# Set labels and title
plt.xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label for Plot A')  # Set label on primary y-axis
ax2.set_ylabel('Y-axis Label for Plot B')  # Set label on secondary y-axis
ax3.set_ylabel('Y-axis Label for Plot C')  # Set label on secondary y-axis
ax4.set_ylabel('Y-axis Label for Plot D')  # Set label on secondary y-axis

# Set equal aspect ratio
plt.gca().set_aspect('equal', adjustable='datalim')

# Add legend with handles from both axes
handles1, labels1 = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
handles3, labels3 = ax3.get_legend_handles_labels()
handles4, labels4 = ax4.get_legend_handles_labels()
plt.legend(handles1 + handles2 + handles3 + handles4, labels1 + labels2 + labels3 + labels4, loc='upper left')  # Combine legends and position

# Show the plot
plt.show()

```
