### Q1)Plot secondary axis in matplotlib with legends


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(pow(x,2))

fig, axis1 = plt.subplots()


axis1.plot(x, y1, 'black', label='y = sin(x)')
axis1.set_xlabel('X data')
axis1.set_ylabel('Primary Y axis', color='black')
axis1.tick_params('y', colors='black')


axis2 = axis1.twinx()


axis2.plot(x, y2, 'r-', label='y = e^(x^2)')
axis2.set_ylabel('Secondary Y axis', color='red')
axis2.tick_params('y', colors='red')


lines1, labels1 = axis1.get_legend_handles_labels()
lines2, labels2 = axis2.get_legend_handles_labels()
axis1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
plt.show()

```

 ### Q2)Use lambda operations on the dataframes

```python

import pandas as pd

# Creating a DataFrame
df = pd.DataFrame({'A': [1, 2, 3, 4],
                   'B': [5, 6, 7, 8]})

df['A_squared'] = df['A'].apply(lambda x: x**2)

print(df)
print()

df['C'] = df.apply(lambda row: row['A'] + row['B'], axis=1)

print(df)
print()

filtered_df = df[df['A'].apply(lambda x: x > 2)]

print(filtered_df)

```

### Q3)Code to check anagrams


```python

def check_anagrams(str1, str2):
    if sorted(str1) == sorted(str2):
        print(f"Both strings {str1} and {str2} are anagrams")
    else:
        print(f"Both strings {str1} and {str2} are not anagrams")
str1="skill"
str2='kills'
str3="cpp"
str4="java"
check_anagrams(str1,str2)
check_anagrams(str3,str4)
```

### 4)Regex example in python


```python
import re

# Example 1: Matching a Pattern
text = "Python is a powerful programming language"
pattern = r'\b\w{6}\b'  
matches = re.findall(pattern, text)
print(matches)
# Example 2: Substituting Patterns
text = "Hello, World! Today is 2023-05-14."
pattern = r'\d{4}-\d{2}-\d{2}'  
replacement = "DATE"
formatted_text = re.sub(pattern, replacement, text)
print(formatted_text)  

# Example 3: Splitting Text
text = "apple,banana,orange,grape"
pattern = r','
tokens = re.split(pattern, text)
print(tokens) 

```

### 5)Explain the types of merges in git

### 1. Fast-Forward Merge:

A Fast-Forward Merge is a type of merge where the main branch has not changed since we created the feature branch. In this case, Git can simply move the main branch forward to point to the latest commit on your feature branch.

### 2. Recursive Merge:

In Recursive merge, after you branch and make some commits, there are some new original commits on the ‘master‘. So, when it’s time to merge, git recurses over the branch and creates a new merge commit. The merge commit continues to have two parents.

### 3.OURS Merge:

This merge strategy resolves any number of heads, but the resulting tree of the merge is always that of the current branch head, effectively ignoring all changes from all other branches. It is meant to be used to supersede the old development history of side branches.

### 4.Octopus Merge:

Octopus Merge strategy resolves cases with more than two heads but refuses to do a complex merge that needs manual resolution. It is primarily meant to be used for bundling topic branch heads together. This is the default merge strategy when pulling or merging more than one branch.

### 5.Resolve Merge:

This strategy can only resolve two heads (i.e. the current branch and another branch you pulled from) using a 3-way merge algorithm. It tries to carefully detect criss-cross merge ambiguities and is considered generally safe and fast

### 6.Subtree Merge:

This is a modified recursive strategy. When merging trees A and B, if B corresponds to a subtree of A, B is first adjusted to match the tree structure of A, instead of reading the trees at the same level. This adjustment is also done to the common ancestor tree.
